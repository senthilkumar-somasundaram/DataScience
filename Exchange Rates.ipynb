# Visualisation of Exchange rates Between 1999 and 2021

## Source: [European Central Bank](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020)


The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

The focus of the project will be on the exchange rate between the Euro and the American dollar.

First, let's start by reading in the dataset.

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
ex_rate_head  = exchange_rates.head()

# Let's check the first 5 rows 
ex_rate_head

```

```python
# Let's check the last 5 rows 
ex_rate_tail  = exchange_rates.tail()
ex_rate_tail
```

Now using the **DataFrame.info()** method we can learn some basic facts about the dataset:


```python
exchange_rates.info()
```

I'll do some modifications to the current dataset. 
* Rename some columns: 
    - **[US dollar ]** to US_dollar
    - **Period\Unit** to Time
* Change the Time column type to a **datetime** data type.
* Sort the values by Time in **ascending** order.
* Reset the index (and drop the **initial index**).

```python
# Rename some column
exchange_rates.rename(columns={'[US dollar ]':'US_dollar', 
                               'Period\\Unit:':'Time'}, 
                      inplace=True)
                               
```

```python
# Let's have a quick look to the changes
exchange_rates
```

```python
# Change the Time column to a **datetime** data type.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
# Sort the values by Time in ascending order.
exchange_rates.sort_values('Time', inplace=True)
```

```python
# Reset the index (and drop the initial index).
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
# Let's isolate the **Time** and the **US_dollar** columns. Assign them to a different variable named **euro_to_dollar**
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar
```

```python
# Let's run the Series.value_counts() method on the US_dollar column, and see if you notice anything wrong.
euro_to_dollar['US_dollar'].value_counts()

```

We can see there are **62 (-)** at the very top

We are dropping all the rows where the ( **-** ) character appears in the US_dollar column

```python
# Leave all values different to ( - )
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# Double check they have been dropped
euro_to_dollar['US_dollar'].value_counts() 
```

```python
# Convert the US_dollar column to a float data type.
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

Now, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.





```python
import matplotlib.pyplot as plt

# Enables Jupyter to display graphs
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

I will calculate the rolling means for the US_dollar column using a moving window of 30 days. Also, I will add the calculated rolling means to a new column named **rolling_mean**

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

euro_to_dollar
```

Let's figure out which of the following story ideas for our data we want to tell:

* We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
* We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2016 and 2009 for comparison. We can use a line plot.
* We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.



## Storytelling Data Visualization Examples
### Financial Crisis Example




```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]

financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2007
                   ) & (euro_to_dollar.Time.dt.year <= 2008)]
# Notice that on the financial_crisis_7_8 when we call for the Time 
# we didn't use the ['Time'] notation
```

We can easily identify what was the highest rate using either **financial_crisis** or the **financial_crisis-7_8** as the data we are looking for is included in both and should be the same.

```python
financial_crisis['US_dollar'].max()
```

```python
financial_crisis_7_8['US_dollar'].max()
```

We can see above that independently of the two (financial_crisis or the financial_crisis-7_8) we use the answer is the same.

Let's zoom in the **financial_crisis_7_8** figure to see it in better detail

```python
plt.figure(figsize=(13, 6))
plt.plot(financial_crisis_7_8['Time'],
         financial_crisis_7_8['US_dollar'])
plt.show()
```

Let's try to see how did the Euro perform in the last three US presincy periods.

We can check the following [Examples](https://matplotlib.org/stable/gallery/index.html) to see what we can do.

We are going to do the 

Steps:
   * We are going to show a row with threee columns one for each period (a subplot line wih three plots on it)
   * Then we will show one more row with the highest value on each period as a bar and the time when it happenned.
   * We can show the average value as a line through all the three bars
    
    
**Separating each presidencial period** 

```python
# Let's recall the euro_to_dollar data

euro_to_dollar
```

```python
# Now, let's get them separate for each period. George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)).

euro_to_dollar_bush = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001
                   ) & (euro_to_dollar['Time'].dt.year < 2009)]

euro_to_dollar_obama  = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2009
                   ) & (euro_to_dollar['Time'].dt.year < 2017)]

euro_to_dollar_trump  = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2017
                   ) & (euro_to_dollar['Time'].dt.year < 2021)]


plt.figure(figsize=(13, 6))
plt.plot(euro_to_dollar_bush['Time'],
         euro_to_dollar_bush['US_dollar'])
plt.show()

bush_max =  euro_to_dollar_bush['US_dollar'].max()
print("Bush Administration Highest Rate:", bush_max)

```

```python

plt.figure(figsize=(13, 6))
plt.plot(euro_to_dollar_obama['Time'],
         euro_to_dollar_obama['US_dollar'])
plt.show()

obama_max =  euro_to_dollar_obama['US_dollar'].max()
print("Obama Administration Highest Rate:", obama_max)
```

```python
plt.figure(figsize=(13, 6))
plt.plot(euro_to_dollar_trump['Time'],
         euro_to_dollar_trump['US_dollar'])
plt.show()

trump_max =  euro_to_dollar_trump['US_dollar'].max()
print("Trump Administration Highest Rate:", trump_max)

```

Also, let's get the exchage rate average during the last three US precidency periods using the mean and the three highest values 


```python
from statistics import mean

last_three_precidency_periods  = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
avg = mean(last_three_precidency_periods['US_dollar'])
avg
```

We can see that the y and x ticks are different on each chart. We are going to join all of them  in a chart  and use the bars to mark the highest value on each period. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Using the information obtain (highest rates on each precidency period)
data = {'George W. Bush (2001-2009)':bush_max,
        'Barack Obama (2009-2017)':obama_max,
        'Donald Trump (2017-2021)':trump_max}

group_data = list(data.values())
group_names = list(data.keys())

def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]+ 0.03,y[i], ha = 'center',color='black')

        
# Main function

if __name__ == '__main__':
    
    # Define data
    x = group_names
    y = group_data

    fig, ax = plt.subplots(figsize=(8, 5.5))
    
    # Set background color of the plot
    ax.set_facecolor('#EFF7FF')


    ### Remove the x tick labels
    ax.set_yticklabels([])

    ### Remove the grid
    ax.grid(b=False)


    # Vertical bars
    ax.bar(group_names, group_data, alpha=0.4, width=0.1)

    #ax.axline(y=0.5, c= 'red', alpha= 0.1,  linewidth = 1.0, xmin=0.1, xmax=0.9)
    ax.axhline(y=1.22, xmin=0.0, xmax=1.0, color='r', alpha= 0.4, linewidth = 1.0)
    
     # Call function
    addtext(x, y)   
    
    ax.text(-0.29, 1.22, '1.22',
        color='black', alpha=1)
    
    ### Adding a title and a subtitle
    ax.text(-0.5, 2.04, 'The average EURO-USD rate under the last three US presidents was 1.22',
             fontsize=20, weight='bold')
    ax.text(-0.5, 1.9, '''Highest EURO-USD exchange rates under the last three US presidents''',
            fontsize=16)
    
    ### Adding a signature
    ax.text(-0.55, -0.25, '©DATAQUEST' + ' '*133 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)
   
```

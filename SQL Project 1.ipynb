#Here are the descriptions for some of the columns:

name      — the name of the country.
area      — the country's total area (both land and water).
area_land  — the country's land area in square kilometers.
area_water — the country's waterarea in square kilometers.
population — the country's population.
population_growth— the country's population growth as a percentage.
birth_rate — the country's birth rate, or the number of births per year per 1,000 people.
death_rate — the country's death rate, or the number of death per year per 1,000 people.


```python
%%sql
SELECT   MIN(POPULATION) as min_population
       , MAX(POPULATION) as max_population
       , MIN(POPULATION_GROWTH) as min_pop_growth
       , MAX(POPULATION_GROWTH) as max_pop_growth
FROM FACTS
```

```python
%%sql
select * from facts where population in (select min(population) from facts)
```

```python
%%sql
select * from facts where population in (select max(population) from facts)
```

```python
%%sql
SELECT   MIN(POPULATION) as min_population
       , MAX(POPULATION) as max_population
       , MIN(POPULATION_GROWTH) as min_pop_growth
       , MAX(POPULATION_GROWTH) as max_pop_growth
FROM FACTS
where name != 'World'
```

```python
%%sql
select   AVG(population) as avg_population
       , AVG(area) as avg_area
    from facts
    
```

```python
%%sql
Select * from facts where population > 
  (select AVG(population) from facts)
    
```

```python
%%sql
Select * from facts where area > 
  (select AVG(area) from facts)
```

```python
%%sql

Select * from facts
 where  populations > (select avg(population) from facts)
    and area        < (select avg(area) from facts)
```
